```python
benign['type']='benign'
m_u['type']='attack'
g_c['type']='attack'
g_j['type']='attack'
g_s['type']='attack'
g_t['type']='attack'
g_u['type']='attack'
m_a['type']='attack'
m_sc['type']='attack'
m_sy['type']='attack'
m_u_p['type']='attack'
```

```python
data=pd.concat([benign,m_u,g_c,g_j,g_s,g_t,g_u,m_a,m_sc,m_sy,m_u_p],
               axis=0, sort=False, ignore_index=True)
```

```python
data.head()
#data.groupby('type')['type'].count()
```

```python
x = data.iloc[:,:-1].values
y = data.iloc[:,-1].values
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
y = labelencoder.fit_transform(y)
```

```python
#standardize numerical columns
# def standardize(df,col):
#     df[col]= (df[col]-df[col].mean())/df[col].std()
# data_st=data.copy()
# for i in (data_st.iloc[:,:-1].columns):
#     standardize (data_st,i)
# data_st.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2,shuffle=True)
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, matthews_corrcoef, f1_score, precision_score, recall_score
```

```python
from keras import Sequential
from keras.layers import Dense

classifier = Sequential()
```

```python
#First Hidden Layer
classifier.add(Dense(64, activation='relu', kernel_initializer='random_normal', input_dim=115))
#Second  Hidden Layer
classifier.add(Dense(32, activation='relu', kernel_initializer='random_normal'))

classifier.add(Dense(16, activation='relu', kernel_initializer='random_normal'))

#Output Layer
classifier.add(Dense(1, activation='sigmoid', kernel_initializer='random_normal'))
```

```python
#Compiling the neural network
classifier.compile(optimizer ='adam',loss='binary_crossentropy', metrics =['accuracy'])
classifier.summary()
```

```python
#Fitting the data to the training dataset
classifier.fit(X_train,y_train, validation_data=(X_test,y_test), epochs=1)
```

```python
y_pred=classifier.predict(X_test)
y_pred =(y_pred>0.5)
```

```python
print(f"Accuracy   : {accuracy_score(y_test, y_pred)}")
print(f"Precision  : {precision_score(y_test, y_pred)}")
print(f"F1-Measure : {f1_score(y_test, y_pred)}")
print(f"MCC        : {matthews_corrcoef(y_test, y_pred)}")
print(f"Recall     : {recall_score(y_test, y_pred)}")
```

```python
print(classification_report(y_test,y_pred))
```

```python
print(confusion_matrix(y_test, y_pred))
```
